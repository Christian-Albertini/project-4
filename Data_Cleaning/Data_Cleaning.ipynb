```python
import pandas as pd
import numpy as np
import random
```

```python
all_star_data=pd.read_csv("Resources/All_Star_Player_Stats.csv")
all_data=pd.read_csv("Resources/Combined_Player_Stats.csv")
```

```python
all_data.head(15)
```

```python
all_star_data.head()
```

```python
def RandomLeague():
    return random.choice(['AL','NL'])

all_data.loc[(all_data.Lg=='AL,NL'),'Lg'] = RandomLeague()
all_data.head(25)
```

```python
all_data_cleaned=all_data.drop(columns=['Age', 'Team', 'GIDP', 'HBP', 'SH', 'SF', 'IBB', 'Pos'])
all_data_cleaned.head()
```

```python
all_star_data_cleaned=all_star_data.drop(columns=['Age', 'Team', 'GIDP', 'HBP', 'SH', 'SF', 'IBB', 'Pos'])
all_star_data_cleaned.head()
```

```python
all_data_cleaned=all_data_cleaned.rename(columns={"Lg":"League",
                                                  "G":"Games_Played",
                                                  "AS":"All-Star",
                                                  "PA":"Plate_Appearances",
                                                  "AB":"At_Bats",
                                                  "R":"Runs",
                                                  "H":"Hits",
                                                  "1B":"Single",
                                                  "2B":"Double",
                                                  "3B":"Triple",
                                                  "HR":"Home_Runs",
                                                  "RBI":"Runs_Batted_In",
                                                  "SB": "Stolen_Bases",
                                                  "CS":"Caught_Stealing",
                                                  "BB":"Walks",
                                                  "SO":"Strikeouts",
                                                  "BA":"Batting_Average",
                                                  "OBP":"On_Base_Percentage",
                                                  "SLG":"Slugging_Percentage",
                                                  "OPS":"On_Base_Plus_Slugging_(OPS)",
                                                  "OPS+":"On_Base_Plus_Slugging_Plus_(OPS+)",
                                                  "TB":"Total_Bases"
                                                  })
all_data_cleaned.head()
```

```python
all_star_data_cleaned=all_star_data_cleaned.rename(columns={"Lg":"League",
                                                  "G":"Games_Played",
                                                  "AS":"All-Star",
                                                  "PA":"Plate_Appearances",
                                                  "AB":"At_Bats",
                                                  "R":"Runs",
                                                  "H":"Hits",
                                                  "1B":"Single",
                                                  "2B":"Double",
                                                  "3B":"Triple",
                                                  "HR":"Home_Run",
                                                  "RBI":"Runs_Batted_In",
                                                  "SB": "Stolen_Bases",
                                                  "CS":"Caught_Stealing",
                                                  "BB":"Walks",
                                                  "SO":"Strikeouts",
                                                  "BA":"Batting_Average",
                                                  "OBP":"On_Base_Percentage",
                                                  "SLG":"Slugging_Percentage",
                                                  "OPS":"On_Base_Plus_Slugging_(OPS)",
                                                  "OPS+":"On_Base_Plus_Slugging_Plus_(OPS+)",
                                                  "TB":"Total_Bases"
                                                  })
all_star_data_cleaned.head()
```

```python
all_data_cleaned=all_data_cleaned.drop(columns=["Player"])
all_star_data_cleaned=all_star_data_cleaned.drop(columns=["Player"])
all_star_data_cleaned.head()
```

```python
all_data_cleaned=all_data_cleaned.dropna()
all_star_data_cleaned=all_star_data_cleaned.dropna()
```

```python
all_data_cleaned=all_data_cleaned.drop_duplicates()
all_star_data_cleaned=all_star_data_cleaned.drop_duplicates()
```

```python
all_star_data_cleaned.head()
```

```python
all_data_cleaned.to_csv("Resources/Combined_Player_Stats_Cleaned_2.csv", index=False)
all_star_data_cleaned.to_csv("Resources/All_Star_Player_Stats_Cleaned_2.csv", index=False)
```
