```python
import pandas as pd
import numpy as np
```

```python
all_data=pd.read_csv("../Resources/Combined_Player_Stats.csv")
standings=pd.read_csv("../Resources/mlb_standings_2013_2023.csv")
```

```python
standings['Team'] = standings['Team'].replace('LAL', 'LAA')
```

```python
condition = all_data['Team'].str.len() <= 3

all_data = all_data[condition]

all_data.head()
```

```python
# Merge all_data with standings
merged_df = pd.merge(all_data, standings[['Team', 'Season', 'Standings']], on=['Team', 'Season'], how='left')
```

```python
merged_df.to_csv("../Resources/Merged_Player_Stats.csv", index=False)
```

```python
all_data=pd.read_csv("../Resources/Merged_Player_Stats.csv")
war=pd.read_csv("../Resources/WAR.csv")
```

```python
# Merge the result with war
merged_df = pd.merge(all_data, war, on=['Player', 'Season'], how='left')
merged_df.head()
```

```python
merged_df = merged_df.drop(columns=['Age', 'OPS', 'TB', 'GIDP', 'HBP', 'SH', 'SF', 'IBB', 'Team_y', 'Rk'])
```

```python
merged_df.head()
```

```python
merged_df = merged_df.rename(columns={
    'Lg': 'League',
    'G': 'Games_Played',
    'AS' : 'All-Star',
    'PA' : 'Plate_Appearances',
    'AB' : 'At_Bats', 
    'R' : 'Runs',
    'H' : 'Hits',
    '1B' : 'Single',
    '2B' : 'Double',
    '3B' : 'Triple',
    'HR' : 'Home_Run',
    'RBI' : 'Runs_Batted_In',
    'SB' : 'Stolen_Bases',
    'CS' : 'Caught_Stealing',
    'BB' : 'Walks',
    'SO' : 'Strikeouts',
    'BA' : 'Batting_Average',
    'OBP' : 'On_Base_Percentage',
    'SLG' : 'Slugging_Percentage',
    'OPS+' : 'On_Base_Plus_Slugging_Plus_(OPS+)',
    'Team_x': 'Team'
})
```

```python
merged_df = merged_df.drop(columns=['Team', 'Single', 'Double', 'Triple', 'Caught_Stealing', 'League'])
```

```python
merged_df.to_csv("../Resources/Merged_Stats_Final.csv")
```
