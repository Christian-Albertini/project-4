```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Use raw URLs from GitHub
combined_url = 'https://raw.githubusercontent.com/Christian-Albertini/project-4/main/Resources/Merged_Stats_Normal.csv'
final_url = 'https://raw.githubusercontent.com/Christian-Albertini/project-4/main/Resources/Merged_Stats_Final_Pos.csv'

# Read the CSV files
combined_df = pd.read_csv(combined_url)
final_df = pd.read_csv(final_url)
```

```python
# Remove freatures that are not needed
combined_df = combined_df.drop(columns=['Player', 'Games_Played', 'Plate_Appearances', 'At_Bats', 'Runs_Batted_In', 'Stolen_Bases', 'Walks', 'Strikeouts', 'Batting_Average', 'On_Base_Percentage', 'Slugging_Percentage', 'Pos'])
final_df = final_df.drop(columns=['Player', 'Games_Played', 'Plate_Appearances', 'At_Bats', 'Runs_Batted_In', 'Stolen_Bases', 'Walks', 'Strikeouts', 'Batting_Average', 'On_Base_Percentage', 'Slugging_Percentage', 'Pos'])
```

```python
# Set the style
plt.style.use('default')
sns.set(style='darkgrid')

# Create the violin plot
fig, ax = plt.subplots(figsize=(10, 6))

sns.violinplot(
    x='All-Star',
    y='PIE_Score',
    data=combined_df,
    hue='All-Star',
    palette={0: 'orange', 1: 'blue'},
    split=True
)

# Customize the plot
ax.set_ylabel('PIE Score')
ax.set_title('PIE Scores by All-Star Status')

# Show the plot
plt.show()
```

```python
# Set the style
sns.set(style='darkgrid')

# Create the density plot
plt.figure(figsize=(10, 6))

sns.kdeplot(
    data=final_df,
    x='Standings',
    hue='All-Star',
    fill=True,
    palette={0: 'orange', 1: 'blue'},
    alpha=0.5
)

# Customize the plot
plt.title('Standings by All-Star Status', fontsize=16)
plt.xlabel('Standings')
plt.ylabel('Density')
plt.legend(title='All-Star Status', labels=['Non-All-Star', 'All-Star'])

# Show the plot
plt.show()
```

```python
# Pairwise scatter plot matrix
sns.pairplot(combined_df, hue='All-Star', palette={0: 'orange', 1: 'blue'})
plt.show()
```
